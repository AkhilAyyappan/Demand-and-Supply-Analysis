```python
import pandas as pd
import plotly.express as pl
import plotly.graph_objects as gp
import plotly.io as pio
pio.templates.default="plotly_white"
data=pd.read_csv('rides.csv')
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
data=data.dropna()
```

```python
demand=data["Riders Active Per Hour"]

supply=data["Drivers Active Per Hour"]
figure=pl.scatter(data,x="Riders Active Per Hour",y="Drivers Active Per Hour",title="demand and supply analysis",trendline="ols")
figure.update_layout(xaxis_title="RIDERS ACTIVE(demand)",yaxis_title="Cab availability(supply)")
figure.show()
```

```python
avgdemnd=data["Riders Active Per Hour"].mean()
avgsupply=data["Drivers Active Per Hour"].mean()
print(avgdemnd,avgsupply)
ptchngedemand=(max(data["Riders Active Per Hour"])-min(data["Riders Active Per Hour"]))/avgdemnd *100
ptchangesupply=(max(data["Drivers Active Per Hour"])-min(data["Drivers Active Per Hour"]))/avgsupply *100
print(ptchngedemand)
print(ptchangesupply)
print("elasticity")
print(ptchngedemand/ptchangesupply)
```

```python
data["supply ratio"]=(data["Rides Completed"]/data["Drivers Active Per Hour"])
print(data.head())
```

```python
fig = gp.Figure()
fig.add_trace(gp.Scatter(x=data['Drivers Active Per Hour'],
                         y=data['supply ratio'], mode='markers'))
fig.update_layout(
    title='Supply Ratio vs. Driver Activity',
    xaxis_title='Driver Activity (Drivers Active Per Hour)',
    yaxis_title='Supply Ratio (Rides Completed per Driver Active per Hour)'
)
fig.show()
```
